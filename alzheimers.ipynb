```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('alzheimers_disease_data.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x=df.drop(['PatientID','DoctorInCharge','Diagnosis'],axis='columns')
```

```python
x
```

```python
y=df['Diagnosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
```

```python
xgb.fit(x_train,y_train)
```

```python
# accuracy score on the training data
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
X_train_prediction = xgb.predict(x_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
print("Accuracy is : ",training_data_accuracy)
```

```python
y_pred=xgb.predict(x_test)
```

```python
xgb.score(x_test,y_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=xgb.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
import pickle

pickle.dump(xgb,open('alzheimers_model.sav','wb'))
```

```python
loaded_model=pickle.load(open('alzheimers_model.sav','rb'))
```

```python
input_data = (75,0,0,1,18.77,0,13.72,4.64,8.34,4.21,0,0,0,0,0,0,117,63,151.38,69.62,77.34,210.57,10.13,3.40,0,0,4.5,1,0,0,0,1)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = loaded_model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The person does not have Alzheimers')
else:
  print('The person has Alzheimers')
```
